## Customer Personality Analysis using Python

### Step 1: Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
```

### Step 2: Load Dataset

```python
df = pd.read_csv("customer_personality.csv",sep="\t")
```

```python
df.head()
```

### Step 3: Dataset Information

```python
df.info()
```

### Step 4: Check Missing Values

```python
df.isnull().sum()
```

### Step 5: Remove Missing Values

```python
df = df.dropna()
df
```

### Step 6: Check Duplicate Records

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
df
```

### Step 7: Univariate Analysis

### Year of Birth Distribution

```python
plt.figure(figsize=(8,5))
plt.hist(df["Year_Birth"],bins = 20)
plt.title("Year of Birth Distribution")
plt.xlabel("Year of Birth")
plt.ylabel("Count")
plt.show()
```

### Income Distribution

```python
plt.figure(figsize=(8,5))
plt.hist(df["Income"],bins=30)
plt.title("Income Distribution")
plt.xlabel("Income")
plt.ylabel("Frequency")
plt.show()
```

### Education

```python
sns.countplot(x = "Education", data = df)
plt.xticks(rotation = 45)
plt.show()
```

### Marital Status

```python
sns.countplot(x = "Marital_Status", data = df)
plt.xticks(rotation = 45)
plt.show()
```

### Step 8: Bivariate Analysis

### Income vs Spending

```python
sns.scatterplot(x = "Income", y = "MntWines", data = df)
plt.title("Income vs Wine Spending")
plt.xlabel("Income")
plt.ylabel("Wine Spending")
plt.show()
```

### Income by Education

```python
plt.figure(figsize=(8,5))
sns.boxplot(x = "Education", y = "Income", data = df)
plt.xticks(rotation = 45)
plt.show()
```

### Step 9: Correlation Heatmap

```python
plt.figure(figsize = (12,8))
numeric_df = df.select_dtypes(include= ['number'])
sns.heatmap(numeric_df.corr(),cmap = "coolwarm", annot = True)
plt.title("Correlation Heatmap")
plt.show()
```

### Step 10: Top Products Purchased

```python
products = ["MntWines","MntFruits","MntMeatProducts","MntFishProducts","MntSweetProducts","MntGoldProds"]
df[products].sum().plot(kind = "bar")
plt.title("Total Spending by Product")
plt.xlabel("Products")
plt.ylabel("Total Spending")
plt.xticks(rotation=45)
plt.show()
```

### Step 11: Insights

- Most customers belong to the **Graduation** education category.
- The **Married** category has the highest number of customers, followed by **Together**.
- Customer income is concentrated between approximately **20,000 and 80,000**, with a few high-income outliers.
- Customers with higher income generally spend more on **wine products**, indicating a positive relationship between income and wine spending.
- **Wine products (MntWines)** have the highest total spending among all product categories.
- **Meat products (MntMeatProducts)** are the second-highest spending category.
- The correlation heatmap shows a positive relationship between **income and product spending**, especially for wine and meat products.
